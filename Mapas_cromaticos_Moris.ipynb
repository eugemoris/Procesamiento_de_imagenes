```python
from PIL import Image,ImageFilter  
import matplotlib.pyplot as plt
import numpy as np
#inline
```

```python
from skimage import data

#obtengo imagen para trabajar
moon = data.moon()
img = Image.fromarray(moon)
```

```python
plt.imshow(moon)
print('Shape: ', np.asarray(img).shape)
```

```python
#@title Choose a palettes. { run: "auto" }
img = img.convert("P", palette=Image.ADAPTIVE, colors=256)
def change_palletes(img,palette):
  plt.contour(img, cmap = palette, levels = list(range(np.min(img), np.max(img), 1)))
  cbar = plt.colorbar()
  plt.imshow(img)

palette = 'magma' #@param ["binary","flag","prism", "twilight", "gist_ncar", "magma","binary"]
change_palletes(img,palette)
```
